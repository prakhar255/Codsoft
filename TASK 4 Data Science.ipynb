# SALES PREDICTION USING PYTHON

```python
# IMPORT THE LIBRARIES
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# LOAD THE DATASET NAME ADVERTISING
file_path=r"C:\Users\Prakhar Khandelwal\Desktop\Internship\Codsoft\task 4\advertising.csv"
data = pd.read_csv(file_path)
```

```python
# DATA VISUALIZATION USING SCATTER PLOT
sns.scatterplot(x=data['TV'],y=data['Sales'],color='b')
plt.title("TV vs Sales")
plt.show()
```

```python
# CORRELATION MATRIX

plt.figure(figsize=(10, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# DISTRIBUTION PLOTS FOR DIFFRENT VARIABLES
plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
sns.histplot(data['TV'], kde=True, color='red')
plt.title('Distribution of TV Spending')

plt.subplot(2, 2, 2)
sns.histplot(data['Newspaper'], kde=True, color='yellow')
plt.title('Distribution of Newspaper Spending')

plt.subplot(2, 2, 3)
sns.histplot(data['Radio'], kde=True, color='lightgreen')
plt.title('Distribution of Radio Spending')

plt.subplot(2, 2, 4)
sns.histplot(data['Sales'], kde=True, color='purple')
plt.title('Distribution of Sales')

plt.tight_layout()
plt.show()
```

```python
# DEFINE THE DATASET
X=data[['TV','Newspaper','Radio']]
Y=data['Sales']
```

```python
# SPLIT THE DATASET
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
# INITIALIZE THE LINEAR REGRESSION MODEL
model=LinearRegression()
```

```python
# TRAIN THE DATASET
model.fit(X_train,Y_train)
```

```python
# MAKE PREDICTION ON THE TEST DATA
pred=model.predict(X_test)
```

```python
# CALCULATE THE MEAN SQUARED ERROR AND R SQUARED VALUE
mae=mean_absolute_error(Y_test,pred)
r2=r2_score(Y_test,pred)
```

```python
# MODEL EVALUATION
print(mae)
print(r2)
```

```python
# PLOT PREDICTION VS ACTUAL SALES
plt.figure(figsize=(10, 6))
plt.scatter(Y_test, pred, color='blue', alpha=0.5)
plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], color='red', linestyle='--')
plt.xlabel('Actual Sales')
plt.ylabel('Predicted Sales')
plt.title('Actual Sales vs Predicted Sales')
plt.show()
```
